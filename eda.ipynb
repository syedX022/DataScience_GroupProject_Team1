```python
# Plot correlation heatmap
corr_matrix = df.select_dtypes(include=['number']).corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()x
```

**Observations from Data Exploration:**
*   The dataset contains house sale data with 1460 entries and 81 features.

*   Initial inspection shows a mix of numerical and categorical features.

*   Several features have missing values, notably `LotFrontage`, `Alley`, `MasVnrType`, `MasVnrArea`, basement-related columns, `FireplaceQu`, garage-related columns, `PoolQC`, `Fence`, and `MiscFeature`.

 *   Histograms revealed varied distributions for numerical features.

 *   Boxplots indicated the presence of outliers in many numerical features.

 *   The correlation analysis of numerical features showed that `OverallQual` has the highest positive correlation with the `SalePrice`, followed by `GrLivArea` and `GarageCars`.
